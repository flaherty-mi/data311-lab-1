```python
# Meghan Flaherty: DATA311
# lab 1 NOAA- notebook 3 
# compare fall weather in belllingham & portland 
# based on temp, humidity, rainfall, & windspeed
#
# worked with Arvind Ramesh for all notebooks in the lab.
#    all lines of code were written together, so all basic logic and technique will be the same
#    some variable names and formats for how the code is written may be different between our submissions
import pandas as pd

bellingham = pd.read_csv('https://facultyweb.cs.wwu.edu/~hutchib2/spr22/data311/lab1/WA_Bellingham.csv')
portland = pd.read_csv('https://facultyweb.cs.wwu.edu/~hutchib2/spr22/data311/lab1/OR_Portland.csv')
```

```python
# isolate columns from the data frames needed for comparison
bham = bellingham[['DATE','REPORT_TYPE','HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']][bellingham['HourlyPrecipitation']!='T'].dropna()
port = portland[['DATE','REPORT_TYPE','HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']][portland['HourlyPrecipitation']!='T'].dropna()
```

```python
# Bellingham September
Bsep = bham[bham['DATE'].str.contains('2020-09')].dropna()
Bsep = Bsep[Bsep['REPORT_TYPE']=='FM-15']

# replace 's' values with '' 
Bsep['HourlyPrecipitation'] = Bsep['HourlyPrecipitation'].str.replace("s", "")
Bsep['HourlyDryBulbTemperature'] = Bsep['HourlyDryBulbTemperature'].str.replace("s", "")

# convert following columns into numeric type
Bsep[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']] = Bsep[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']].apply(pd.to_numeric)

# get mean values of following columns 
Bsep_DBT = Bsep['HourlyDryBulbTemperature'].mean()
Bsep_WBT = Bsep['HourlyWetBulbTemperature'].mean()
Bsep_WS = Bsep['HourlyWindSpeed'].mean()
Bsep_P = Bsep['HourlyPrecipitation'].mean()
Bsep_RH = Bsep['HourlyRelativeHumidity'].mean()

# Portland September
Psep = port[port['DATE'].str.contains('2020-09')].dropna()
Psep = Psep[Psep['REPORT_TYPE']=='FM-15']

# replace 's' values with '' 
Psep['HourlyPrecipitation'] = Psep['HourlyPrecipitation'].str.replace("s", "")

# convert following columns into numeric type
Psep[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']] = Psep[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']].apply(pd.to_numeric)

# get mean values of following columns 
Psep_DBT = Psep['HourlyDryBulbTemperature'].mean()
Psep_WBT = Psep['HourlyWetBulbTemperature'].mean()
Psep_WS = Psep['HourlyWindSpeed'].mean()
Psep_P = Psep['HourlyPrecipitation'].mean()
Psep_RH = Psep['HourlyRelativeHumidity'].mean()
```

```python
# Bellingham October
Boct = bham[bham['DATE'].str.contains('2020-10')].dropna()
Boct = Boct[Boct['REPORT_TYPE']=='FM-15']

# replace 's' values with '' 
Boct['HourlyPrecipitation'] = Boct['HourlyPrecipitation'].str.replace("s", "")
Boct['HourlyDryBulbTemperature'] = Boct['HourlyDryBulbTemperature'].str.replace("s", "")

# convert following columns into numeric type
Boct[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']] = Boct[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']].apply(pd.to_numeric)

# get mean values of following columns 
Boct_DBT = Boct['HourlyDryBulbTemperature'].mean()
Boct_WBT = Boct['HourlyWetBulbTemperature'].mean()
Boct_WS = Boct['HourlyWindSpeed'].mean()
Boct_P = Boct['HourlyPrecipitation'].mean()
Boct_RH = Boct['HourlyRelativeHumidity'].mean()

# Portland October
Poct = port[port['DATE'].str.contains('2020-10')].dropna()
Poct = Poct[Poct['REPORT_TYPE']=='FM-15']
# replace 's' values with '' 
Poct['HourlyPrecipitation'] = Poct['HourlyPrecipitation'].str.replace("s", "")

# convert following columns into numeric type
Poct[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']] = Poct[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']].apply(pd.to_numeric)

# get mean values of following columns 
Poct_DBT = Poct['HourlyDryBulbTemperature'].mean()
Poct_WBT = Poct['HourlyWetBulbTemperature'].mean()
Poct_WS = Poct['HourlyWindSpeed'].mean()
Poct_P = Poct['HourlyPrecipitation'].mean()
Poct_RH = Poct['HourlyRelativeHumidity'].mean()
```

```python
# Bellingham November
Bnov = bham[bham['DATE'].str.contains('2020-11')].dropna()
Bnov = Bnov[Bnov['REPORT_TYPE']=='FM-15']

# replace 's' values with '' 
Bnov['HourlyPrecipitation'] = Bnov['HourlyPrecipitation'].str.replace("s", "")
Bnov['HourlyDryBulbTemperature'] = Bnov['HourlyDryBulbTemperature'].str.replace("s", "")

# convert following columns into numeric type
Bnov[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']] = Bnov[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']].apply(pd.to_numeric)

# get mean values of following columns 
Bnov_DBT = Bnov['HourlyDryBulbTemperature'].mean()
Bnov_WBT = Bnov['HourlyWetBulbTemperature'].mean()
Bnov_WS = Bnov['HourlyWindSpeed'].mean()
Bnov_P = Bnov['HourlyPrecipitation'].mean()
Bnov_RH = Bnov['HourlyRelativeHumidity'].mean()

# Portland November
Pnov = port[port['DATE'].str.contains('2020-11')].dropna()
Pnov = Pnov[Pnov['REPORT_TYPE']=='FM-15']

# replace 's' values with '' 
Pnov['HourlyPrecipitation'] = Pnov['HourlyPrecipitation'].str.replace("s", "")

# convert following columns into numeric type
Pnov[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']] = Pnov[['HourlyDryBulbTemperature','HourlyWetBulbTemperature','HourlyWindSpeed', 'HourlyPrecipitation', 'HourlyRelativeHumidity']].apply(pd.to_numeric)

# get mean values of following columns 
Pnov_DBT = Pnov['HourlyDryBulbTemperature'].mean()
Pnov_WBT = Pnov['HourlyWetBulbTemperature'].mean()
Pnov_WS = Pnov['HourlyWindSpeed'].mean()
Pnov_P = Pnov['HourlyPrecipitation'].mean()
Pnov_RH = Pnov['HourlyRelativeHumidity'].mean()
```
